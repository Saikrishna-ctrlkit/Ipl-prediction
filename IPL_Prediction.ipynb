```python
cat_col =df.select_dtypes(include='object').columns
print(f"Categorical columns: {list(cat_col)}")
```

```python
df['runs_target'] = df['runs_target'].fillna(0)

# win_outcome NaN = no result match — fill with 'no_result'
df['win_outcome'] = df['win_outcome'].fillna('no_result')

# Drop columns with too many nulls — not useful for ML
cols_to_drop = [
    'extra_type', 'wicket_kind', 'player_out', 'fielders',
    'review_batter', 'team_reviewed', 'review_decision', 'umpire',
    'superover_winner', 'result_type', 'method', 'power_surge_start',
    'new_batter', 'next_batter', 'Unnamed: 0'
]
df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)

print(f"Shape after cleaning: {df.shape}")
print("Remaining nulls:", df.isnull().sum()[df.isnull().sum() > 0].to_dict())
```

```python
# Drop the rows with single nulls (only 1 row affected)
cols_with_one_null = ['over', 'ball', 'ball_no', 'batter', 'bat_pos',
                      'runs_batter', 'balls_faced', 'bowler', 'valid_ball',
                      'runs_extras', 'runs_total', 'runs_bowler', 'runs_not_boundary',
                      'non_striker', 'non_striker_pos', 'umpires_call',
                      'player_of_match', 'match_won_by', 'toss_winner',
                      'toss_decision', 'venue', 'city']

df.dropna(subset=[c for c in cols_with_one_null if c in df.columns], inplace=True)

print(f"Shape after dropna: {df.shape}")
print("Nulls remaining:", df.isnull().sum()[df.isnull().sum() > 0].to_dict())
```

```python
# STEP 2: FEATURE ENGINEERING

# Sort by match, innings, over, ball
df = df.sort_values(['match_id', 'innings', 'over', 'ball']).reset_index(drop=True)

# Current Run Rate
df['current_run_rate'] = np.where(
    df['team_balls'] > 0,
    df['team_runs'] / (df['team_balls'] / 6),
    0
)

# Balls remaining
df['balls_remaining'] = (120 - df['team_balls']).clip(lower=0)
df['overs_remaining'] = df['balls_remaining'] / 6

# Runs needed (2nd innings)
df['runs_needed'] = df['runs_target'] - df['team_runs']

# Required Run Rate
df['required_run_rate'] = np.where(
    (df['innings'] == 2) & (df['overs_remaining'] > 0),
    df['runs_needed'] / df['overs_remaining'],
    0
).clip(0)

# Pressure Index
df['pressure_index'] = np.where(
    df['current_run_rate'] > 0,
    df['required_run_rate'] / df['current_run_rate'],
    0
)

# Extra features
df['wickets_remaining'] = 10 - df['team_wicket']
df['is_powerplay']      = (df['over'] <= 6).astype(int)
df['is_death_over']     = (df['over'] >= 16).astype(int)
df['batting_won_toss']  = (df['batting_team'] == df['toss_winner']).astype(int)
df['toss_bat']          = (df['toss_decision'] == 'bat').astype(int)

print("Step 2 done!")
print(df[['team_runs','team_balls','current_run_rate',
          'required_run_rate','balls_remaining','wickets_remaining']].head())
```

```python
# STEP 3: ENCODE CATEGORICAL VARIABLES
from sklearn.preprocessing import LabelEncoder

cat_cols = ['batting_team', 'bowling_team', 'batter', 'bowler',
            'venue', 'city', 'toss_winner', 'toss_decision',
            'non_striker', 'stage']

le_dict = {}
for col in cat_cols:
    if col in df.columns:
        le = LabelEncoder()
        df[col + '_enc'] = le.fit_transform(df[col].astype(str))
        le_dict[col] = le

print("Step 3 done!")
print("Encoded:", [c+'_enc' for c in cat_cols if c in df.columns])
```

```python
# STEP 4: TRAIN/TEST SPLIT BY MATCH
import numpy as np

match_ids = df['match_id'].unique()
np.random.seed(42)
np.random.shuffle(match_ids)

split = int(0.8 * len(match_ids))
train_ids = match_ids[:split]
test_ids  = match_ids[split:]

train_df = df[df['match_id'].isin(train_ids)].copy()
test_df  = df[df['match_id'].isin(test_ids)].copy()

print(f"Train: {len(train_ids)} matches | {len(train_df)} rows")
print(f"Test : {len(test_ids)} matches | {len(test_df)} rows")
```

```python
# STEP 5A: SCORE PREDICTION (1st innings)
import xgboost as xgb
from sklearn.metrics import mean_absolute_error

# Get final score per match
final_scores = (
    df[df['innings'] == 1]
    .groupby('match_id')['team_runs']
    .max()
    .reset_index()
    .rename(columns={'team_runs': 'final_score'})
)

train_s = train_df[train_df['innings'] == 1].merge(final_scores, on='match_id')
test_s  = test_df[test_df['innings'] == 1].merge(final_scores, on='match_id')

score_features = [
    'team_runs', 'team_balls', 'team_wicket', 'wickets_remaining',
    'current_run_rate', 'balls_remaining', 'over', 'ball',
    'bat_pos', 'batter_runs', 'batter_balls',
    'is_powerplay', 'is_death_over', 'batting_won_toss', 'toss_bat',
    'batting_team_enc', 'bowling_team_enc', 'batter_enc',
    'bowler_enc', 'venue_enc', 'city_enc', 'month', 'day'
]

X_train_s = train_s[score_features]
y_train_s = train_s['final_score']
X_test_s  = test_s[score_features]
y_test_s  = test_s['final_score']

score_model = xgb.XGBRegressor(
    n_estimators=300, max_depth=6,
    learning_rate=0.05, subsample=0.8,
    colsample_bytree=0.8, random_state=42, verbosity=0
)
score_model.fit(X_train_s, y_train_s)

mae = mean_absolute_error(y_test_s, score_model.predict(X_test_s))
print(f"✅ Score Model MAE: {mae:.2f} runs")
```

```python
# STEP 5B: MATCH WINNER PREDICTION (2nd innings)
from sklearn.metrics import accuracy_score, classification_report

train_m = train_df[train_df['innings'] == 2].copy()
test_m  = test_df[test_df['innings'] == 2].copy()

# Target: 1 if batting team wins, 0 if bowling team wins
train_m['batting_team_wins'] = (train_m['win_outcome'] == train_m['batting_team']).astype(int)
test_m['batting_team_wins']  = (test_m['win_outcome'] == test_m['batting_team']).astype(int)

match_features = [
    'team_runs', 'team_balls', 'team_wicket', 'wickets_remaining',
    'current_run_rate', 'required_run_rate', 'balls_remaining',
    'runs_target', 'runs_needed', 'pressure_index',
    'over', 'ball', 'bat_pos', 'batter_runs', 'batter_balls',
    'is_powerplay', 'is_death_over', 'batting_won_toss', 'toss_bat',
    'batting_team_enc', 'bowling_team_enc', 'batter_enc',
    'bowler_enc', 'venue_enc', 'city_enc', 'month', 'day'
]

X_train_m = train_m[match_features]
y_train_m = train_m['batting_team_wins']
X_test_m  = test_m[match_features]
y_test_m  = test_m['batting_team_wins']

match_model = xgb.XGBClassifier(
    n_estimators=300, max_depth=5,
    learning_rate=0.05, subsample=0.8,
    colsample_bytree=0.8, eval_metric='logloss',
    random_state=42, verbosity=0
)
match_model.fit(X_train_m, y_train_m)

acc = accuracy_score(y_test_m, match_model.predict(X_test_m))
print(f"✅ Match Winner Accuracy: {acc*100:.2f}%")
# Fix: Since y_test_m contains only class 0, we explicitly tell classification_report
# to only consider label 0 and provide a single target name.
print(classification_report(y_test_m, match_model.predict(X_test_m),
      labels=[0], target_names=['Bowling team wins']))
```

```python
# STEP 6: LIVE PREDICTION FUNCTIONS

def safe_encode(le, val):
    val = str(val)
    return list(le.classes_).index(val) if val in le.classes_ else -1

def predict_score(batting_team, bowling_team, batter, bowler,
                  venue, city, team_runs, team_balls, team_wicket,
                  bat_pos, batter_runs, batter_balls, over, ball,
                  toss_winner, toss_decision, month=4, day=15):

    crr = team_runs / (team_balls/6) if team_balls > 0 else 0
    balls_remaining = max(0, 120 - team_balls)

    row = pd.DataFrame([{
        'team_runs': team_runs, 'team_balls': team_balls,
        'team_wicket': team_wicket, 'wickets_remaining': 10 - team_wicket,
        'current_run_rate': crr, 'balls_remaining': balls_remaining,
        'over': over, 'ball': ball, 'bat_pos': bat_pos,
        'batter_runs': batter_runs, 'batter_balls': batter_balls,
        'is_powerplay': int(over<=6), 'is_death_over': int(over>=16),
        'batting_won_toss': int(batting_team==toss_winner),
        'toss_bat': int(toss_decision=='bat'),
        'batting_team_enc': safe_encode(le_dict['batting_team'], batting_team),
        'bowling_team_enc': safe_encode(le_dict['bowling_team'], bowling_team),
        'batter_enc': safe_encode(le_dict['batter'], batter),
        'bowler_enc': safe_encode(le_dict['bowler'], bowler),
        'venue_enc': safe_encode(le_dict['venue'], venue),
        'city_enc': safe_encode(le_dict['city'], city),
        'month': month, 'day': day
    }])

    pred = score_model.predict(row[score_features])[0]
    print(f"\n{'='*40}")
    print(f"  {batting_team} vs {bowling_team}")
    print(f"  Score: {team_runs}/{team_wicket} in over {over}.{ball}")
    print(f"  ✅ Predicted Final Score: {int(pred)} runs")
    print(f"{'='*40}")


def predict_winner(batting_team, bowling_team, batter, bowler,
                   venue, city, team_runs, team_balls, team_wicket,
                   bat_pos, batter_runs, batter_balls, over, ball,
                   runs_target, toss_winner, toss_decision, month=4, day=15):

    crr = team_runs / (team_balls/6) if team_balls > 0 else 0
    balls_remaining = max(0, 120 - team_balls)
    overs_remaining = balls_remaining / 6
    runs_needed = runs_target - team_runs
    rrr = runs_needed / overs_remaining if overs_remaining > 0 else 0
    pressure = rrr / crr if crr > 0 else 0

    row = pd.DataFrame([{
        'team_runs': team_runs, 'team_balls': team_balls,
        'team_wicket': team_wicket, 'wickets_remaining': 10 - team_wicket,
        'current_run_rate': crr, 'required_run_rate': max(0,rrr),
        'balls_remaining': balls_remaining, 'runs_target': runs_target,
        'runs_needed': runs_needed, 'pressure_index': pressure,
        'over': over, 'ball': ball, 'bat_pos': bat_pos,
        'batter_runs': batter_runs, 'batter_balls': batter_balls,
        'is_powerplay': int(over<=6), 'is_death_over': int(over>=16),
        'batting_won_toss': int(batting_team==toss_winner),
        'toss_bat': int(toss_decision=='bat'),
        'batting_team_enc': safe_encode(le_dict['batting_team'], batting_team),
        'bowling_team_enc': safe_encode(le_dict['bowling_team'], bowling_team),
        'batter_enc': safe_encode(le_dict['batter'], batter),
        'bowler_enc': safe_encode(le_dict['bowler'], bowler),
        'venue_enc': safe_encode(le_dict['venue'], venue),
        'city_enc': safe_encode(le_dict['city'], city),
        'month': month, 'day': day
    }])

    proba = match_model.predict_proba(row[match_features])[0]
    print(f"\n{'='*40}")
    print(f"  {batting_team} vs {bowling_team}")
    print(f"  Chasing {runs_target} | {team_runs}/{team_wicket} in over {over}.{ball}")
    print(f"  Need {int(runs_needed)} off {balls_remaining} balls")
    print(f"  CRR: {crr:.2f} | RRR: {rrr:.2f}")
    print(f"  🏏 {batting_team}: {proba[1]*100:.1f}%")
    print(f"  🎳 {bowling_team}: {proba[0]*100:.1f}%")
    print(f"{'='*40}")


# --- TEST IT ---
predict_score(
    batting_team='Mumbai Indians',
    bowling_team='Chennai Super Kings',
    batter='RG Sharma', bowler='R Jadeja',
    venue='Wankhede Stadium', city='Mumbai',
    team_runs=85, team_balls=60, team_wicket=2,
    bat_pos=1, batter_runs=45, batter_balls=32,
    over=10, ball=1,
    toss_winner='Mumbai Indians', toss_decision='bat'
)

predict_winner(
    batting_team='Chennai Super Kings',
    bowling_team='Mumbai Indians',
    batter='MS Dhoni', bowler='JJ Bumrah',
    venue='Wankhede Stadium', city='Mumbai',
    team_runs=120, team_balls=90, team_wicket=5,
    bat_pos=5, batter_runs=20, batter_balls=15,
    over=15, ball=1, runs_target=175,
    toss_winner='Mumbai Indians', toss_decision='bat'
)
```
